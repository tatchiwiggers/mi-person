```python
import pandas as pd

df =  pd.read_csv('../mi-person/data/go_emotions.csv')
# df.head(14)
# df.iloc[14]
```

# Positive

```python
df = df.loc[df['example_very_unclear'] == False]

df['positive'] = 0
df.loc[df[["admiration","amusement","approval","caring","curiosity",\
    "desire","excitement","gratitude","joy","love","optimism","relief"]].sum(axis=1) == 1, "positive" ] = 1

df['negative'] = 0
df.loc[df[["anger","annoyance","confusion","disappointment","disapproval","disgust","embarrassment",\
    "fear","grief","nervousness","pride","remorse","sadness"]].sum(axis=1) == 1, "negative" ] = 1

df_test = df[['text', 'positive', 'neutral', 'negative']]
df_test.to_csv('df_test.csv')
```
